#### What should we predict?
Let's predict the sport by sex, age, height or weight, and noc or team

Label:
* sport - categorical

Predictors:
* sex - categorical, binary
* age - numeric, in years
* height - numeric, in centimeters. correlated with weight
* weight - numeric, in kilogram
* noc - categorical

```python
# how many unique sports are there to classify

ath_clean['sport'].nunique()
```

```python
# because there are so many sports, let's narrow down how many we classify
# Athletics, Swimming, and Gymnastics have the most instances. Let's go with these.

ath_clean.groupby('sport').count().sort_values('ath_id', ascending=False)[['ath_id']].head(5)
```

```python
# filter for just these sports

ath_filter = ath_clean[ath_clean['sport'].isin(['Athletics', 'Swimming', 'Gymnastics'])]
```

```python
# how small is our dataset now?

ath_filter.shape
```

#### Time to choose an algorithm
Let's first try k-nearest neighbors

https://scikit-learn.org/stable/modules/neighbors.html

https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html#sklearn.neighbors.KNeighborsClassifier

* classification
* lazy learner
* no assumptions about the unlying distribution

```python
# convert the numerical variables to categorical

le = LabelEncoder()
```

```python
ath_filter['sex_num'] = le.fit_transform(ath_filter['sex'])

ath_filter['noc_num'] = le.fit_transform(ath_filter['noc'])

ath_filter['sport_num'] = le.fit_transform(ath_filter['sport'])

ath_filter['team_num'] = le.fit_transform(ath_filter['team'])
```

```python
# divide data into training (2/3) and testing (1/3)

X_train, X_test, y_train, y_test = train_test_split(ath_filter[['sex_num','age','height','noc_num']], ath_filter['sport_num'], 
                                                    test_size=0.33)
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train)
```

```python
predictions = neigh.predict(X_test)
```

```python
neigh.score(X_test, y_test)
```

## Interpret

```python
def get_metrics(y_test, y_predicted):  
    # true positives / (true positives+false positives)
    precision = precision_score(y_test, y_predicted, pos_label=None,
                                    average='weighted')             
    # true positives / (true positives + false negatives)
    recall = recall_score(y_test, y_predicted, pos_label=None,
                              average='weighted')
    
    # harmonic mean of precision and recall
    f1 = f1_score(y_test, y_predicted, pos_label=None, average='weighted')
    
    # true positives + true negatives/ total
    accuracy = accuracy_score(y_test, y_predicted)
    return accuracy, precision, recall, f1

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.winter):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title, fontsize=20)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, fontsize=20)
    plt.yticks(tick_marks, classes, fontsize=20)
    
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.

    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), horizontalalignment="center", 
                 color="white" if cm[i, j] < thresh else "black", fontsize=20)
    
    plt.tight_layout()
    plt.ylabel('True label', fontsize=20)
    plt.xlabel('Predicted label', fontsize=20)

    return plt
```

```python
accuracy, precision, recall, f1 = get_metrics(y_test, predictions)
print("accuracy = %.2f, precision = %.2f, recall = %.2f, f1 = %.2f" % (accuracy, precision, recall, f1))
```

```python
cm = confusion_matrix(y_test, predictions)

plt.figure()
plot_confusion_matrix(cm, classes=['0','1','2'], normalize=True,
                      title='Normalized confusion matrix')

plt.show()
```
